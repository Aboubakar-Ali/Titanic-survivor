```python
import pandas as pd

# Charger les données
df = pd.read_csv("../Data/train.csv")

# Gérer les valeurs manquantes
# Imputer les valeurs manquantes pour l'âge avec la médiane
age_median = df['Age'].median()
df['Age'].fillna(age_median, inplace=True)

# Imputer les valeurs manquantes pour le point d'embarquement avec le mode
embark_mode = df['Embarked'].mode()[0]
df['Embarked'].fillna(embark_mode, inplace=True)

# Encodage des variables catégorielles
# Encoder la colonne 'Sex' en utilisant l'encodage binaire
df['Sex'] = df['Sex'].map({'female': 0, 'male': 1})

# Encoder la colonne 'Embarked' en utilisant l'encodage ordinal
df['Embarked'] = df['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})

# Normalisation des données
# Normaliser la colonne 'Age' en utilisant la normalisation min-max
age_min = df['Age'].min()
age_max = df['Age'].max()
df['Age'] = (df['Age'] - age_min) / (age_max - age_min)

df.to_csv('preprocessed_data.csv', index=False)
df


```
